```python
class ClusterMethodList(object) :
    def get_cluster_instance(self, argument,input_data,X):
        method_name = str(argument).lower()+ '_cluster_algorithm'
        method = getattr(self,method_name,lambda : "Invalid Clustering method")
        return method(input_data,X)
    
    def kmeans_cluster_algorithm(self,input_data,X):
        km = cluster.KMeans(n_clusters =input_data['n_clusters'] )
        km_fit = km.fit_predict(X)
        km_score = km.score(X)
        print("The KMeans Cluster Algorithm Score:",km_score)
        km_cross = cross_val_score(km, ML2019)
        print("The KMeans Cluster algorithm Cross Value Score:",km_cross.mean())
    
        return km_fit
   
    def meanshift_cluster_algorithm(self,input_data,X):
        ms = cluster.MeanShift(bandwidth=input_data['bandwidth'])
        ms_fit = ms.fit_predict(X)
        
        return  ms_fit
    
    def minibatchkmeans_cluster_algorithm(self,input_data,X):
        two_means = cluster.MiniBatchKMeans(n_clusters=input_data['n_clusters'])
        two_means_fit =two_means.fit_predict(X)
        two_means_score = two_means.score(X)
        print("The Minibatchkmeans Cluster Algorithm Score:",two_means_score)
        return two_means_fit
   
    def dbscan_cluster_algorithm(self,input_data,X):
        db = cluster.DBSCAN(eps=input_data['eps'])
        db_fit = db.fit_predict(X)
      
        return db_fit
    
    def spectral_cluster_algorithm(self,input_data,X):
        sp = cluster.SpectralClustering(n_clusters=input_data['n_clusters'])
        sp_fit = sp.fit_predict(X)
        
        return sp_fit
   
    def affinitypropagation_cluster_algorithm(self,input_data,X):
        affinity_propagation =  cluster.AffinityPropagation(damping=input_data['damping'], preference=input_data['preference'])
        affinity_propagation.fit(X)
        
        return affinity_propagation.predict(X)
       
    
    def birch_cluster_algorithm(self,input_data,X):
        birch = cluster.Birch(n_clusters=input_data['n_clusters'])
        birch_fit = birch.fit_predict(X)
       
        return birch_fit
   
    def gaussian_mixture_cluster_algorithm(self,input_data,X):
        gmm = mixture.GaussianMixture( n_components=input_data['n_clusters'], covariance_type='full')
        gmm.fit(X)
        gmm_score = gmm.score(X)
        print("The Gaussian_mixture_cluster_algorithm Score:",gmm_score)
        return  gmm.predict(X)
    
        
```

```python
def startClusteringProcess(list_cluster_method,input_data,no_columns,data_set):
    fig,ax = plt.subplots(no_rows,no_columns, figsize=(10,10)) 
    cluster_list = ClusterMethodList()
    i = 0
    j=0
    for cl in list_cluster_method :
        cluster_result = cluster_list.get_cluster_instance(cl,input_data,data_set)
        #convert cluster result array to DataFrame
        data_set[cl] = pd.DataFrame(cluster_result)
        ax[i,j].scatter(data_set.iloc[:, 4], data_set.iloc[:, 5],  c=cluster_result)
        ax[i,j].set_title(cl+" Cluster Result")
        j=j+1
        if( j % no_columns == 0) :
            j= 0
            i=i+1
    plt.subplots_adjust(bottom=-0.5, top=1.5)
    plt.show()
```

```python
list_cluster_method = ['KMeans',"MeanShift","MiniBatchKmeans","DBScan","Spectral","AffinityPropagation","Birch","Gaussian_Mixture"]
# For Graph display 
no_columns = 2
no_rows = 4
# NOT all algorithms require this parameter
n_clusters= 100
bandwidth = 0.1 
# eps for DBSCAN
eps = 0.3
## Damping and perference for Affinity Propagation clustering method
damping = 0.9
preference = -200
input_data = {'n_clusters' :  n_clusters, 'eps' : eps,'bandwidth' : bandwidth, 'damping' : damping, 'preference' : preference}
```

```python
startClusteringProcess(list_cluster_method,input_data,no_columns,ML2019)
```

Underfitting emerges when the internal structure of the data does not sufficiently execute a mathematical model or machine learning algorithm. It happens when the model or algorithm doesn't properly match the data. If the model or algorithm has less variance but high bias, underfitting occurs. ## Cross Validation
Cross-validation is a method for validating the effectiveness of the proposed machine learning model by training it on the given input dataset and testing it on the previously unknown given input sub-set. There is still the requirement to assess the model's stability. It ensures based on the training dataset just on the training dataset, we cannot fit our model. We have reserved a specific sample of the set of data for this purpose, that was not part of the training dataset. After that, before deployment, we evaluate our model on that sample and so this full phase is cross-validated.The cross validation was only implementable to K-means Clustering in my dataset and for other algorithms it was just zero or not applicable.

## Confusion Matrix
When coming to the confusion matrix, It is a much easier way to measure a classifier's output. The basic principle is to count the number of times the class J occurrences are classified as class K.  The occurrences in the expected class are represented in each row of the matrix, although each column contains the occurrences in the actual class. It is a rare type of contingency table with two dimensions which are actual and predicted and similar sets are present in both class J, K in dimensions.
